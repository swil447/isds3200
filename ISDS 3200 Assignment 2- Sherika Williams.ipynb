```python
#Assessment for Uniqueness of traffic crash data by checking whether or not there is duplication of information 

def Check_Duplication(Data_Frame):
    new_DF=Data_Frame[Data_Frame.duplicated(keep=False)]
    return new_DF

Check_Duplication(Traffic_Crash_DF)

#There is no duplication of information in this dataset. All information is unique.



```

```python
#Assessment for Uniqueness of traffic signal data by checking whether or not there is duplication of information 

Check_Duplication(Traffic_Signal_DF)

#There is no duplication of information in this dataset. All information is unique.
```

```python
#Assessment for Uniqueness of road closure data by checking whether or not there is duplication of information 

Check_Duplication(Road_Closure_DF)

#There is no duplication of information in this dataset. All information is unique.
```
